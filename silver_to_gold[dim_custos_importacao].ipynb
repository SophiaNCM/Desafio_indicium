# Lendo a tabela da camada silver

```python
#Lendo a tabela
tbl_custos_importacao= spark.read.table('lh_nautical.silver_lh_nautical.tbl_custos_importacao')
tbl_custos_importacao.display()
```

# Criando a tabela dim_custos_importacao

```python
%sql
-- Informando o schema e o catalogo
USE CATALOG lh_nautical;
USE SCHEMA gold_lh_nautical;
-- Criando uma tabela identica a tabela tbl_custos_importacao da camada silver na camada gold
CREATE TABLE dim_custos_importacao AS
SELECT product_id, product_name, category, usd_price, start_date FROM lh_nautical.silver_lh_nautical.tbl_custos_importacao
```

# Conferindo a nova tabela 

```python
%sql
-- Conferindo a tabela
SELECT * FROM dim_custos_importacao
```

```python
%sql
-- Verificando o range de datas
SELECT DISTINCT MIN(start_date), MAX(start_date) FROM dim_custos_importacao
```
