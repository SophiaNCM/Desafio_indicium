# Lendo a tabela da camada silver

```python
# Lendo a tabela
fatos_vendas= spark.read.table('lh_nautical.silver_lh_nautical.tbl_vendas')
fatos_vendas.display()
```

# Criando a tabela fatos_vendas


```python
%sql
--Informando o schema e o catalogo
USE CATALOG lh_nautical;
USE SCHEMA gold_lh_nautical;
-- Criando uma tabela identica a tabela da camada silver na gold
CREATE TABLE fatos_vendas AS
SELECT id,id_client,id_product,qtd,total,sale_date FROM lh_nautical.silver_lh_nautical.tbl_vendas
```

# Conferindo a tabela

```python
%sql
--Conferindo se a tabela foi criada
SELECT * FROM fatos_vendas
```
