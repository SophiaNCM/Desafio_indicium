# Modificando a tabela

### - Adicionando as colunas id_cities e id_states e criando a tabela

```python
from pyspark.sql.functions import col, when
#Conferindo as tabelas que vamos utilizar
dim_estados = spark.read.table('lh_nautical.gold_lh_nautical.dim_estados')
dim_clientes = spark.read.table('lh_nautical.silver_lh_nautical.tbl_clientes')
dim_cidades = spark.read.table('lh_nautical.gold_lh_nautical.dim_cidades')
dim_estados.display()
dim_clientes.display()
dim_cidades.display()
```

```python
%sql
-- Criando a tabela dim_clientes com os ids das cidades e estados ao inves dos nomes
USE CATALOG lh_nautical;
USE SCHEMA gold_lh_nautical;
CREATE TABLE dim_clientes AS
SELECT  lh_nautical.silver_lh_nautical.tbl_clientes.code, lh_nautical.silver_lh_nautical.tbl_clientes.email, lh_nautical.silver_lh_nautical.tbl_clientes.full_name, lh_nautical.gold_lh_nautical.dim_cidades.id AS id_cities, dim_estados.id AS id_states 
FROM lh_nautical.gold_lh_nautical.dim_estados 
INNER JOIN lh_nautical.silver_lh_nautical.tbl_clientes ON dim_estados.abbreviations = lh_nautical.silver_lh_nautical.tbl_clientes.states 
INNER JOIN lh_nautical.gold_lh_nautical.dim_cidades ON tbl_clientes.cities = lh_nautical.gold_lh_nautical.dim_cidades.cities
```

### - Conferindo a tabela dim_clientes
