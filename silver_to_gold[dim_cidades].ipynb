# Criando uma tabela dim_ciade e colando a coluna cities da tabela tbl_clientes

```python
%sql
-- Informando o schema e catalogo
USE CATALOG lh_nautical;
USE SCHEMA gold_lh_nautical;
-- Criando uma tabela dim_cidades com os dados da coluna cities da tabela tbl_clientes
CREATE TABLE IF NOT EXISTS dim_cidades AS SELECT DISTINCT cities FROM silver_lh_nautical.tbl_clientes 

```

# Verificando a tabela

```python
%sql
-- Verificando a tabela
SELECT * FROM dim_cidades
```

# Adicionando a coluna id

```python
%sql
-- Alterando a tabela para adicionar uma coluna id
ALTER TABLE dim_cidades ADD COLUMN id INT
```

# Populando a tabela id

```python
from pyspark.sql.functions import *
cidades = spark.read.table('gold_lh_nautical.dim_cidades')
cidades.display()
```

```python
from pyspark.sql import Window
from pyspark.sql.functions import *
# Populando o campo id 
cidades = cidades.withColumn('id', monotonically_increasing_id())
w = Window.orderBy('id')
cidades = cidades.withColumn('id', row_number().over(w))
cidades.display()
# Salvando a modificação
cidades.write.mode('overwrite').saveAsTable('dim_cidades')

```
