```python

tbl_vendas = spark.read.format("csv").option("header", "true").option("delimiter", ",").load("/Volumes/lh_nautical/datalake/datas/vendas_2023_2024.csv")
tbl_vendas.display()
```

# - Parte 1

```python
import pyspark.sql.functions as F

# Quantidade total de linhas
num_linhas = tbl_vendas.count()
#Quantidade total de colunas
num_colunas = len(tbl_vendas.columns)
#Intervalo de datas analisado(data mínima e máxima)
min_data = tbl_vendas.select(F.min("sale_date")).collect()[0][0]
max_data = tbl_vendas.select(F.max("sale_date")).collect()[0][0]

print(f'numeros de linhas: {num_linhas}')
print(f'numeros de linhas: {num_colunas}')
print(f'data mínima: {min_data}')
print(f'data máxima: {max_data}')
```

# - Parte 2

```python
spark.sql('SELECT total FROM tbl_vendas')
```

```python
#Valor mínimo do total
min_valor_total = tbl_vendas.select(F.min("total")).collect()[0][0]
#Valor máximo do total
max_valor_total = tbl_vendas.select(F.max("total")).collect()[0][0]
#Valor médio do total
medio_valor_total = tbl_vendas.select(F.avg("total")).collect()[0][0]

print(f"Valor mínimo do total: {min_valor_total}")
print(f"Valor máximo do total: {max_valor_total}")
print(f"Valor médio do total: {medio_valor_total}")
```
