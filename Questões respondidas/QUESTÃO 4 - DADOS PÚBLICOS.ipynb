```python
%sql
USE CATALOG lh_nautical;
USE SCHEMA gold_lh_nautical;
with fatos_vendas as (
    select *
    FROM fatos_vendas
   
),

importacao_em_real as (
 select 
  product_id,
  product_name as nome_produto,
  avg(dim_cotacao_dolar.value * dim_custos_importacao.usd_price) as preco_medio_br
from dim_custos_importacao 
LEFT JOIN dim_cotacao_dolar 
ON dim_custos_importacao.start_date = dim_cotacao_dolar.date
group by all
),

base as(
SELECT 
fatos_vendas.id_product as id_produtos,
nome_produto,
total as total,
(fatos_vendas.qtd * preco_medio_br) as valor_minimo
from fatos_vendas
LEFT JOIN importacao_em_real
ON fatos_vendas.id_product = importacao_em_real.product_id

)

select nome_produto,id_produtos, sum(total) as receita, sum(valor_minimo) as custo, receita - custo as prejuizo, receita/prejuizo as percentual_perda from base group by id_produtos, nome_produto


```
