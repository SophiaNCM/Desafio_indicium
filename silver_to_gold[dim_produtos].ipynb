# Criando a tabela dim_produtos

```python
%sql
-- Informando o catalogo e schema
USE CATALOG lh_nautical;
USE SCHEMA gold_lh_nautical;
-- Criando a tabela dim_produtos
CREATE TABLE dim_produtos AS
SELECT code, name, monetary_unit, price, actual_category FROM lh_nautical.silver_lh_nautical.tbl_produtos

     
```

# Verificando a tabela

```python
%sql
--Conferindo a tabela
SELECT * FROM dim_produtos
```
