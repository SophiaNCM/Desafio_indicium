#Lendo um csv com os nomes e ufs dos estados brasileiros para criação da tabela dim_estados 

```python
# Lendo arquivo csv
schema = 'id INT,  abbreviations STRING, states STRING'
estados = spark.read.format('csv').option('header','true').schema(schema).load('/Volumes/lh_nautical/datalake/datas/estados_brasileiros.csv')
estados.display()
```

# Acessando o schema Gold

```python
%sql
-- Informando o catalogo 
USE CATALOG lh_nautical;
--Informando o schema
USE gold_lh_nautical
```

### - Salvando a tabela na camada Gold

```python
#Salvando o arquivo como tabela na camada gold
estados.write.mode('overwrite').saveAsTable('gold_lh_nautical.dim_estados')
```

### - Conferindo a tabela

```python
#Conferindo se a tabela foi salva
estados = spark.read.table('gold_lh_nautical.dim_estados')
estados.display()

```
